```python
import numpy as np
import pandas as pd
```

## Data Cleansing

```python
price_data = pd.read_csv('data_medianprice.csv')
price_data.info()
```

```python
price_data.head()
```

```python
education_data = pd.read_csv('data_education.csv')
risk_data = pd.read_csv('data_risks.csv')
```

```python
df1 = pd.merge(price_data, education_data, on = ['County', 'MonthNumeric', 'Year'])
df1.info()
```

```python
etc_data = pd.read_csv('final_df.csv')
df2 = pd.merge(df1, etc_data, on = ['County', 'Year'])
df2.info()
```

```python
etc2_data = pd.read_csv('data.csv')
df3 = pd.merge(df2, etc2_data, on = ['County'])
df3.info()
```

```python
df4 = pd.merge(df3, risk_data, on = ['County', 'MonthNumeric', 'Year'])
df4.info()
```

```python
df4 = df4.drop(columns = ['Unnamed: 0'])
```

```python
df4['Median Price'] = df4['Median Price'].apply(lambda x: x.split('$')[1])
df4.head()
```

```python
df4['Median Price'] = df4['Median Price'].str.replace(',','')
df4["Median Price"] = df4["Median Price"].astype(float)
```

```python
county_type = pd.read_csv('County_Type.csv')
county_type.head()
```

```python
df5 = df4.merge(county_type, on = ['County'])
df5.info()
```

```python
employment_data = pd.read_csv('employment_2010.csv')
walmart_costco_data = pd.read_csv('WalmartCostco.csv')

employment_data.head()
```

```python
walmart_costco_data.head()
```

```python
walmart_costco_data = walmart_costco_data.rename(columns = {'CountyName':'County'})
employment_data['Employment'] = employment_data['Employment'].replace('\t','', regex=True)
```

```python
df6 = df5.merge(walmart_costco_data, on = ['County'])
df7 = df6.merge(employment_data, on = ['County', 'Year'])
df7.info()
```

```python
df7 = df7.rename(columns = {'Employment':'MajorIndustry'})
df7.head(3)
```

```python
df7['Local Tax Rate'] = df7['Local Tax Rate'].str.rstrip('%').astype('float') / 100.0

df7.info()
```

```python
df7 = df7.rename(columns={"Median Price": "MedianPrice",
                          "% of Bachelor's Degree": "PercentageOfBachelorDegree",
                          "Num Airports": "NumAirports",
                          "Local Tax Rate": "LocalTaxRate", 
                          "Violent Crime per 1000 people": "ViolentCrimeRate",
                          "Property Crime per 1000 people": "PropertyCrimeRate",
                          "Grad Rates": "GradRates",
                          "College Enrollment Rate": "CollegeEnrollmentRate",
                          "Avg Household Size": "AvgHouseholdSize",
                          "Avg #ppl <18": "Avg18Younger",
                          "Avg #ppl 18+": "Avg18Older",
                          "National Risk Index": "NationalRiskIndex"})

df7.info()
```

```python
df7.to_csv('final_dataframe.csv')
```

```python
train = df7[df7['Year'] <= 2015]
test = df7[df7['Year'] > 2015]

len(train), len(test)
```

```python
train.to_csv('project_train.csv')
test.to_csv('project_test.csv')
```
